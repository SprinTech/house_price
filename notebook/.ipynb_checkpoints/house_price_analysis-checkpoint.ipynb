## Analyse multivariée des données quantitatives

```python
df_int_float = df.select_dtypes(exclude='object')
df_int_float.columns[:]
```

```python
df_int_float.hist(figsize = (20,20))
```

Parmi les valeurs ayant une distribution proche de SAlePrice, on distingue :

### Liste des 10 valeurs les plus corrélés avec le prix de vente

```python
df.corr()['SalePrice'].sort_values(ascending = False).head(11)
```

En voyant ces corrélations, on peut affirmer :
    - que la qualité du bien et la surface habitable influencent très fortement le prix
    - que le nombre de place disponible dans le garage (et donc la superficie de celui-ci) influence très fortement le prix
    - que l'âge du bien influence le prix de manière modérée 

### Création d'un pairplot

```python
def pairplot_viz(df, y, x):
    sns.pairplot(data=df,
                  y_vars= y,
                  x_vars= x)
    
pairplot_viz(df_int_float, 'SalePrice', df_int_float.columns[:])
```

On observe que les plot des 3 valeurs les plus corrélés montre une progression du positionnement des points dans l'axe y (SalePrice) à mesure que l'on progresse dans l'axe x.

### Création d'une heatmap

```python
df_int_float_corr = df_int_float.corr()

cmap = sns.diverging_palette(h_neg=210, h_pos=350, s=90, l=30, as_cmap=True)

sns.clustermap(df_int_float_corr, linewidths=.5, figsize=(10, 10), cmap=cmap)
plt.title("Correlation between features", 
          weight='bold', 
          fontsize=18)
plt.xticks(weight='bold')
plt.yticks(weight='bold')
plt.show()
```

Parmi les pairs les plus fortement corrélés, on retrouve : 
    - Le prix et la qualité du bien
    - La surface habitable et le nombre total de pièces (hors sous-sol)
    - La superficie du sous-sol et la superficie du premier étage

### Création d'un regplot

```python
columns = ['TotalBsmtSF', '1stFlrSF', '2ndFlrSF', 'GrLivArea', 'FullBath', 'TotRmsAbvGrd', 'GarageCars', 'GarageArea', 'SalePrice']
for col in columns:
    plt.figure(figsize = (7,7))
    sns.regplot(x = col, y = 'SalePrice', data = df_int_float)
```

```python
df_int_float.columns[:]
```

### Création de jointplot

```python
graph = sns.jointplot(data=df_int_float, x= df_int_float.SalePrice, y=df_int_float.GrLivArea,kind = "reg")
r, p = stats.pearsonr(df_int_float.SalePrice, df_int_float.GrLivArea)

phantom, = graph.ax_joint.plot([], [], linestyle="", alpha=0)

graph.ax_joint.legend([phantom],['r={:f}, p={:f}'.format(r,p)])
```
